## Выполним очистку преобразование и очистку данных

```python
# Давайте определим pipeline для очистки всех сообщений
# pipeline выполняет следующие действия: (1) удаляет знаки препинания, (2) удаляет стоп-слова

def message_cleaning(message):
    Test_punc_removed = [char for char in message if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    Test_punc_removed_join_clean = [word for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return Test_punc_removed_join_clean
```

```python
# протестируем недавно добавленную функцию
reviews_df_clean = reviews_df['verified_reviews'].apply(message_cleaning)
```

```python
# очищенная версия
print(reviews_df_clean[3]) 
```

```python
# оригинальная версия
print(reviews_df['verified_reviews'][3])
```

```python
reviews_df_clean
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer = message_cleaning)
reviews_countvectorizer = vectorizer.fit_transform(reviews_df['verified_reviews'])
```

```python
print(vectorizer.get_feature_names())
```

```python
# сначала давайте отбросим колонку
reviews_df.drop(['verified_reviews'], axis=1, inplace=True)
reviews = pd.DataFrame(reviews_countvectorizer.toarray())
```

```python
# Теперь давайте объединим их вместе
reviews_df = pd.concat([reviews_df, reviews], axis=1)
reviews_df
```

```python
# Давайте отбросим целевой признак
X = reviews_df.drop(['feedback'],axis=1)
X
```

```python
y = reviews_df['feedback']
```

## Модель классификатор наивный Байес

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_train = NB_classifier.predict(X_train)
y_predict_train
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot=True)
```

```python
# Прогнозирование результатов тестового набора
y_predict_test = NB_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict_test))
```

## Классификатор логистической регресии

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print('Accuracy {} %'.format( 100 * accuracy_score(y_pred, y_test)))
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_pred))
```
